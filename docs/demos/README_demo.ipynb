```python
# Add the root directory to path. Only needed if SystemFlow not in packages
import sys
sys.path.append("../..")
```

```python
import systemflow as sf

# Define Stocks 
clouds = sf.Stock("clouds", initial_value=500)
lake = sf.Stock("lake", initial_value=50)

# Link the stocks
rain = clouds >> lake

# Create flow logic to the links
condensation = clouds * 0.2

# Set the logic to the flow's 
rain.valve = condensation

# Simulate and display results
system = sf.System(rain, stop=20)
system.simulate()
system.plot()
```

### Saving the figure to images

```python
demo_fig = system.plot()[0]
demo_fig.savefig('../img/readme_fig.png', bbox_inches="tight")
```
